```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head()
```

## Player Count

* Display the total number of players


```python
totalplayers = purchase_data["SN"].count()
totalplayers
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python

totalrevenue = purchase_data["Price"].sum()
totalrevenue

totalpurchases = purchase_data["Purchase ID"].count()
totalpurchases
uniqueitems = len(purchase_data["Item ID"].unique())
uniqueitems

averageprice1 = purchase_data["Price"].mean()
averageprice1

averageprice2 = totalrevenue/totalpurchases
averageprice2

purchase_analysis = pd.DataFrame([{"Number of Unique Items": uniqueitems, "Average Price": averageprice1,
                                      "Number of Purchases": totalpurchases, "Total Revenue": totalrevenue}])
purchase_analysis["Average Price"] = purchase_analysis["Average Price"].map("${:,.2f}".format)
purchase_analysis["Total Revenue"] = purchase_analysis["Total Revenue"].map("${:,.2f}".format)
purchase_analysis

# reorganazing the column order for summary dataframe
org_purchase_analysis = purchase_analysis[["Number of Unique Items","Average Price","Number of Purchases", "Total Revenue" ]]
org_purchase_analysis
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results

* Display a preview of the summary data frame



```python
grpSN = purchase_data.groupby("SN")
grpSN.count()

analysisSPENDOR = pd.DataFrame(grpSN["Purchase ID"].count())

# Get Total purchase value by SN
totalSN = grpSN["Price"].sum()


# Get Average purchase price by SN
avgSN = grpSN["Price"].mean()
tavgSN = avgSN.map("${:,.2f}".format)

# Organize summary Top Spender data, get all columns to organized Data Frame, add needed columns to it
analysisSPENDOR["Average Purchase Price"] = tavgSN
analysisSPENDOR["Total Purchase Value"] = totalSN 


SNpurchased = analysisSPENDOR.rename(columns={"Purchase ID":"Purchase Count"})
TOP5spendors= SNpurchased.sort_values("Total Purchase Value", ascending=False)

ftotal = totalSN.map("${:,.2f}".format)
TOP5spendors["Total Purchase Value"] = ftotal
TOP5spendors.head()
```

## Most Popular Items

* Create a summary data frame to hold the results

* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
grpitem = purchase_data.groupby(["Item ID", "Item Name"])

analysisITEM = pd.DataFrame(grpitem["Purchase ID"].count())
totalITEM = grpitem["Price"].sum()
ftotalITEM = totalITEM.map("${:,.2f}".format)
purchaseprice = grpitem["Price"].mean()
fpurchaseprice = purchaseprice.map("${:,.2f}".format)
analysisITEM["Item Price"] = fpurchaseprice
analysisITEM["Total Purchase Value"] = ftotalITEM
SUMITEM = analysisITEM.rename(columns={"Purchase ID":"Purchase Count"})
TOP5ITEMS = SUMITEM.sort_values("Purchase Count", ascending=False)
TOP5ITEMS.head()
```
